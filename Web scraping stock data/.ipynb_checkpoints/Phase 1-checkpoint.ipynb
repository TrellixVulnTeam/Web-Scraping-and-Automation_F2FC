```python
# Function to create data frame to store in csv file
def dataProcessing(page,underlying_price):
    data_calls = page["optionChain"]["result"][0]["options"][0]["calls"]
    data_puts = page["optionChain"]["result"][0]["options"][0]["puts"]
    #Adding the required columns
    for data in data_calls:
        now = datetime.now()
        data["Ticker"] = symbol
        data["Calls or Puts"] = "Calls"
        data["Today Date"] = now.strftime("%Y-%m-%d")
        data["Time of Retrieval"] = now.strftime("%H:%M:%S")
        expiration_date = datetime.fromtimestamp(data['expiration'])
        data['expiration'] = expiration_date.strftime("%Y-%m-%d")
        lastTrade_date = datetime.fromtimestamp(data['lastTradeDate'])
        data['lastTradeDate'] = lastTrade_date.strftime("%Y-%m-%d %H:%M:%S")
        data['Days left to Expiration'] = (pd.to_datetime(data['expiration']) - pd.to_datetime(data['Today Date'])).days
    #     print(data)

    for data in data_puts:
        now = datetime.now()
        data["Ticker"] = symbol
        data["Calls or Puts"] = "Puts"
        data["Today Date"] = now.strftime("%Y-%m-%d")
        data["Time of Retrieval"] = now.strftime("%H:%M:%S")
        expiration_date = datetime.fromtimestamp(data['expiration'])
        data['expiration'] = expiration_date.strftime("%Y-%m-%d")
        lastTrade_date = datetime.fromtimestamp(data['lastTradeDate'])
        data['lastTradeDate'] = lastTrade_date.strftime("%Y-%m-%d %H:%M:%S")
        data['Days left to Expiration'] = (pd.to_datetime(data['expiration']) - pd.to_datetime(data['Today Date'])).days
        
        
    Columns = [
        'Ticker', 'Calls or Puts', 'Today Date', 'Time of Retrieval',
        'contractSymbol', 'strike', 'currency', 'lastPrice', 'change',
        'percentChange', 'volume', 'openInterest', 'bid', 'ask', 'contractSize',
        'expiration', 'lastTradeDate', 'impliedVolatility', 'inTheMoney','Days left to Expiration'
    ]    
    
    #Creating dataframe of today's data
    df_calls = pd.DataFrame(data_calls,columns=Columns)
    df_puts = pd.DataFrame(data_puts,columns=Columns)
    df = pd.concat([df_calls,df_puts])
    # openInterest less than 10 data is stored
    df = df[df['openInterest'] > 10]
    df['Underlying Price'] = underlying_price
    
    return df
    
```

```python
# Append any more symbols to create them
symbols = ['SPY','GLD','AAPL','NFLX','FB']
url = "https://query2.finance.yahoo.com/v7/finance/options/"

for symbol in symbols:
    try:
        page = requests.get(url+symbol).json()
        print("Data Acquired")
    except:
        print("Couldn't Load data")

    try:
        underlying_page = requests.get('https://finance.yahoo.com/quote/SPY?p=SPY')
        soup = BeautifulSoup(underlying_page.content, 'html.parser')
        spans = soup.find('span',{'class':'Trsdu(0.3s) Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(b)'})
        underlying_price = spans.get_text()    
        print("Got Underlying Price")
    except:
        print("Couldn't fetch Underlying Price data")
        
    try:
        df = dataProcessing(page,underlying_price)
#         print(df.head(3))
    except:
        print("Couldn't get Data in Correct Format")
        
    today = datetime.today().strftime('%Y-%m-%d')
    fname = 'Database/' + symbol + "_" + today + ".csv"
    print(fname)
    if os.path.isfile(fname):
        print("File already exists")
        with open(fname, 'a') as f:
            df.to_csv(f, header=False,index=False)
    else:
        print("New file created")
        df.to_csv(fname,index=False)
    print("\n")        
```
